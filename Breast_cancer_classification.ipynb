## Exploratory data analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#What is the distribution of the target variable diagnosis
# Countplot for target variable 'diagnosis'
sns.countplot(data=df, x='diagnosis', palette='Set1')
plt.title('Distribution of Diagnosis (Malignant vs Benign)')
plt.show()
```

```python
#What are the distributions of key features like radius_mean, perimeter_mean, area_mean, and concavity_mean?
# Distribution plot for radius_mean
# List of key features to plot
key_features = ['radius_mean', 'perimeter_mean', 'area_mean', 'concavity_mean']

# 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Flatten the axes for easier iteration
axes = axes.flatten()

# Loop through key features and plot each in the grid
for i, feature in enumerate(key_features):
    sns.histplot(df[feature], kde=True, bins=30, ax=axes[i], color='blue')
    axes[i].set_title(f'Distribution of {feature}')

# Adjust layout to avoid overlapping plots
plt.tight_layout()

# Show the grid of plots
plt.show()
```

```python
#What are the summary statistics (mean, median, standard deviation) of all numerical features?
# Summary statistics for numerical features
print(df.describe())
```

## Bivariate EDA

```python
#How are radius_mean and perimeter_mean related?
# Scatterplot of radius_mean vs perimeter_mean
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='radius_mean', y='perimeter_mean', hue='diagnosis', palette='Set1')
plt.title('Scatterplot of Radius Mean vs Perimeter Mean by Diagnosis')
plt.show()
```

```python
#What is the correlation between numerical features?
# Correlation matrix

# Convert diagnosis column to numerical: 1 for Malignant, 0 for Benign
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#How does compactness_mean vary between malignant and benign diagnoses?
# Violin plot for compactness_mean by diagnosis
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='diagnosis', y='compactness_mean', palette='Set1')
plt.title('Compactness Mean by Diagnosis')
plt.show()
```

# Multivariate EDA

```python
#How do radius_mean, texture_mean, and area_mean interact across different diagnoses?
# Pairplot of key features, colored by diagnosis
sns.pairplot(df[['radius_mean', 'texture_mean', 'area_mean', 'diagnosis']], hue='diagnosis', palette='Set1')

# Adjust the layout to make space for the title
plt.subplots_adjust(top=0.9)  # Adjust the top margin

# Add the title
plt.suptitle('Pairplot of Key Features by Diagnosis', fontsize=16)

# Show the plot
plt.show()
```

```python
#What are the principal components that explain the variance in the dataset?
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
features = df.drop(['id', 'diagnosis'], axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Plot the PCA result
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['diagnosis'].apply(lambda x: 1 if x == 'M' else 0), cmap='coolwarm')
plt.title('PCA of Breast Cancer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
